# <font color='green'> <b>Exploratory Data Analysis</b><font color='black'>

**Data Set Information:**

This dataset was obtained from UCI Machine Learning Repository. The aim of this problem is to classify adults in two different groups based on their income where group 1 has an income less than USD 50k and group 2 has an income of more than or equal to USD 50k. The data available at hand comes from Census 1994.

**Veri Seti Bilgisi:**

Bu veri seti UCI Machine Learning Repository'den alınmıştır. Bu problemin amacı, gelirlerine göre yetişkinleri iki farklı gruba ayırmaktır. Birinci grup 50.000 USD'den daha az geliri olanları, ikinci grup ise 50.000 USD veya daha fazla geliri olanları içermektedir. Elimizdeki veriler 1994 Nüfus Sayımı'ndan gelmektedir.

**Attribute Information:**

### Outlier Kontrolü

```python
plt.style.use('ggplot')
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()
features = df.select_dtypes(include ="number").columns.drop("Income")
for i, col in enumerate(features):
    sns.boxplot(y=df[col], ax=axes[i], color='skyblue')
    axes[i].set_title(col)
    axes[i].set_ylabel('')
plt.tight_layout()
plt.show()
```

```python
num_cols = [
    "Age", "Fnlwgt", "Net_Capital", "Hours_Per_Week"
]

def outlier_summary(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    return pd.Series({
        "lower_bound": lower,
        "upper_bound": upper,
        "outlier_count": ((df[col] < lower) | (df[col] > upper)).sum()
    })

df_outliers = pd.DataFrame(
    {col: outlier_summary(df, col) for col in num_cols}
).T

df_outliers
```

```python
cols = ["Age", "Hours_Per_Week"]

for col in cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower, upper)
```

```python
num_cols = [
    "Age", "Fnlwgt", "Net_Capital", "Hours_Per_Week"
]

def outlier_summary(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    return pd.Series({
        "lower_bound": lower,
        "upper_bound": upper,
        "outlier_count": ((df[col] < lower) | (df[col] > upper)).sum()
    })

df_outliers = pd.DataFrame(
    {col: outlier_summary(df, col) for col in num_cols}
).T

df_outliers
```

- IQR yöntemi kullanılarak yapılan aykırı değer analizinde, üst sınırın üzerinde kalan gözlemler tespit edilmiştir. Bu değerlerin değişkenin doğal dağılımından mı yoksa veri hatasından mı kaynaklandığı incelenmiş ve modelleme aşamasında winsorization(Sınır değere eşitleme) yöntemi uygulanmıştır.
- Fnlwgt: Her gözlemin nüfusu ne kadar temsil ettiğini gösteren ağırlık (sample weight) olduğu için winsorization uygulamadık. Modelin öğrenmesi için bir feature değil, İstatistiksel dengeleme amaçlı bir feature olduğu için dokunmadık.
- Education_Num, Education değişkeninin sayısal karşılığıdır. O yüzden winsorization uygulamadık.
- Net_Capital, winsorization uygulamadık. Gözlemlerin çok büyük kısmı 0'lardan oluşuyor, 0 olmayan değerler gerçek ve çok bilgi taşıyor.Gelirle güçlü ilişkili bilgiyi tamamen yok etmiş oluyoruz.

# <font color='green'> <b>Implement Logistic Regression</b><font color='black'>

## Train | Test Split

```python
X = df.drop(["Income"], axis = 1)
y = df.Income
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify = y, random_state = 42)
```

## Modelling

### Vanilla Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.pipeline import Pipeline
```

```python
operations = [("scaler", StandardScaler()), ("logistic", LogisticRegression())]

pipe_model = Pipeline(steps = operations)

pipe_model.fit(X_train, y_train)
```

### Prediction

```python
y_pred = pipe_model.predict(X_test) 
y_pred_proba = pipe_model.predict_proba(X_test)
```

```python
test_data = pd.concat([X_test, y_test], axis = 1)
test_data["pred"] = y_pred
test_data["pred_proba"] = y_pred_proba[:,1]
test_data.sample(20)
```

### Model Performance

```python
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
```

```python
ConfusionMatrixDisplay.from_estimator(pipe_model, X_test, y_test);
```

```python
def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```

```python
eval_metric(pipe_model, X_train, y_train, X_test, y_test)
```

- 0 ve 1 sınıfının metriklerindeki büyük fark bana bu verinin unbalanced olduğunu söylüyor. 0 sınıfımızda daha fazla veri var. 

## Cross Validation

```python
from sklearn.model_selection import cross_validate, StratifiedKFold
```

```python
operations = [("scaler", StandardScaler()), ("logistic", LogisticRegression())]

model = Pipeline(steps=operations)

cv = StratifiedKFold(n_splits = 5) # Dengesiz verisetlerinde target sınıflarının oranını koruyarak veriyi böler.

scores = cross_validate(model, X_train, y_train, 
                        scoring = ["precision", "recall", "f1", "accuracy"], 
                        cv = cv, return_train_score = True)

df_scores = pd.DataFrame(scores, index = range(1,6))

df_scores
```

```python
df_scores.mean()[2:]
```

## Cross validate for 0 class

```python
from sklearn.metrics import make_scorer
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score

f1_0 = make_scorer(f1_score, pos_label =0)
precision_0 = make_scorer(precision_score, pos_label =0)
recall_0 = make_scorer(recall_score, pos_label =0)

scoring = {"precision_0":precision_0, "recall_0":recall_0, "f1_0":f1_0}

operations = [("scaler", StandardScaler()), ("logistic", LogisticRegression())]

model = Pipeline(steps=operations)

cv = StratifiedKFold(n_splits=5)

scores = cross_validate(model, X_train, y_train, scoring = scoring, cv = cv, return_train_score=True)
df_scores = pd.DataFrame(scores, index = range(1, 6))
df_scores
```

```python
df_scores.mean()[2:]
```

- 0 sınıfında daha fazla veri olduğu için modelim 0 sınıfını daha iyi öğrenmiş, daha yüksek skorlar aldık. Aşağıda 1 sınıfının skorları mevcuttur.

## GridSearchCV

```python
pipe_model.get_params()
```

```python
from sklearn.model_selection import GridSearchCV

operations = [("scaler", StandardScaler()), ("logistic", LogisticRegression())]

model = Pipeline(steps = operations)

penalty = ["l1", "l2"]
class_weight = ["balanced"]
solver = ["liblinear", "saga"]
C = np.logspace(-2, 2, 10)

param_grid = {
    "logistic__penalty": penalty,
    "logistic__C": C,
    "logistic__class_weight": class_weight,
    "logistic__solver": solver
}

cv = StratifiedKFold(n_splits = 5, shuffle=True, random_state = 42)

grid_model = GridSearchCV(estimator = model,
                         param_grid = param_grid,
                         scoring = {
                                    "precision": "precision",
                                    "recall": "recall",
                                    "f1": "f1"
                                    },
                         refit = "f1",
                         n_jobs = -1,
                         cv = cv,
                         verbose = 1,
                         return_train_score = True)
```

```python
grid_model.fit(X_train, y_train)
```

```python
grid_model.best_params_
```

## Elbow Method

```python
from sklearn.model_selection import cross_validate

test_error_rates = []

for k in range(1, 20):
    operations = [("scaler", StandardScaler()), ("knn", KNeighborsClassifier(n_neighbors = k))]
    knn_pipe_model = Pipeline(steps = operations)
    
    scores = cross_validate(knn_pipe_model, X_train, y_train, scoring = ["accuracy"], cv = 5)
    accuracy_mean = scores["test_accuracy"].mean()
    test_error = 1 - accuracy_mean
    test_error_rates.append(test_error)
```

```python
test_error_rates
```

```python
plt.figure(figsize=(15,8))
plt.plot(range(1,20),
        test_error_rates,
        color = "red",
        marker="o",
        markerfacecolor="yellow",
        markersize=10)

plt.title("Error Rate vs K Value")
plt.xlabel("K_values")
plt.ylabel("Error Rate")

plt.hlines(y = 0.18, xmin = 0, xmax = 20, colors = "b", linestyle = "--") # k = 4 için hata miktarını daha net görelim.
plt.hlines(y = 0.17311456, xmin = 0, xmax = 20, colors = "b", linestyle = "--") # k = 13 için hata miktarını inceleyelim.

plt.show()
```

```python
from sklearn.metrics import make_scorer

accuracy_scores = []
recall_scores = []
precision_scores = []

for k in range(1, 20):
    operations = [("scaler", StandardScaler()), ("knn", KNeighborsClassifier(n_neighbors=k))]
    knn_pipe_model = Pipeline(steps=operations)
    
    scoring = {
        'accuracy': make_scorer(accuracy_score),
        'recall': make_scorer(recall_score),
        'precision': make_scorer(precision_score)
    }
    
    scores = cross_validate(knn_pipe_model, X_train, y_train, scoring=scoring, cv=5)
    
    accuracy_mean = scores["test_accuracy"].mean()
    recall_mean = scores["test_recall"].mean()
    precision_mean = scores["test_precision"].mean()
    
    accuracy_scores.append(accuracy_mean)
    recall_scores.append(recall_mean)
    precision_scores.append(precision_mean)

fig, ax1 = plt.subplots(figsize=(15, 8))

color = 'tab:red'
ax1.set_xlabel('K Values')
ax1.set_ylabel('Accuracy', color=color)
ax1.plot(range(1, 20), accuracy_scores, color=color, marker='o', markerfacecolor='yellow', markersize=10, label='Accuracy')
ax1.tick_params(axis='y', labelcolor=color)
ax1.legend(loc='upper left')

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Recall/Precision', color=color)
ax2.plot(range(1, 20), recall_scores, color='green', marker='s', markerfacecolor='blue', markersize=10, label='Recall')
ax2.plot(range(1, 20), precision_scores, color=color, marker='^', markerfacecolor='orange', markersize=10, label='Precision')
ax2.tick_params(axis='y', labelcolor=color)
ax2.legend(loc='upper right')

plt.title('Metrics vs. K Value')
fig.tight_layout()
plt.show()
```

```python
accuracy_errors = []
recall_errors = []
precision_errors = []

for k in range(1, 20):
    operations = [("scaler", StandardScaler()), ("knn", KNeighborsClassifier(n_neighbors=k))]
    knn_pipe_model = Pipeline(steps=operations)
    
    scoring = {
        'accuracy': make_scorer(accuracy_score),
        'recall': make_scorer(recall_score),
        'precision': make_scorer(precision_score)
    }
    
    scores = cross_validate(knn_pipe_model, X_train, y_train, scoring=scoring, cv=5)
    
    accuracy_mean = scores["test_accuracy"].mean()
    recall_mean = scores["test_recall"].mean()
    precision_mean = scores["test_precision"].mean()
    
    accuracy_errors.append(1 - accuracy_mean)
    recall_errors.append(1 - recall_mean)
    precision_errors.append(1 - precision_mean)

fig, ax1 = plt.subplots(figsize=(15, 8))

color = 'tab:red'
ax1.set_xlabel('K Values')
ax1.set_ylabel('1 - Accuracy', color=color)
ax1.plot(range(1, 20), accuracy_errors, color=color, marker='o', markerfacecolor='yellow', markersize=10, label='1 - Accuracy')
ax1.tick_params(axis='y', labelcolor=color)
ax1.legend(loc='upper left')

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('1 - Recall/1 - Precision', color=color)
ax2.plot(range(1, 20), recall_errors, color='green', marker='s', markerfacecolor='blue', markersize=10, label='1 - Recall')
ax2.plot(range(1, 20), precision_errors, color=color, marker='^', markerfacecolor='orange', markersize=10, label='1 - Precision')
ax2.tick_params(axis='y', labelcolor=color)
ax2.legend(loc='upper right')

plt.title('Error Metrics vs. K Value')
fig.tight_layout()
plt.show()
```

### Overfitting control for k values

```python
test_error_rates = []
train_error_rates = []

for k in range(1, 20):
    operations = [("scaler", StandardScaler()), ("knn", KNeighborsClassifier(n_neighbors = k))]
    knn_pipe_model = Pipeline(steps = operations)
    
    scores = cross_validate(knn_pipe_model, X_train, y_train, scoring = ["accuracy"], cv = 5, return_train_score = True)
    
    accuracy_test_mean = scores["test_accuracy"].mean()
    accuracy_train_mean = scores["train_accuracy"].mean()

    
    test_error = 1 - accuracy_test_mean
    train_error = 1 - accuracy_train_mean
    
    test_error_rates.append(test_error)
    train_error_rates.append(train_error)
```

```python
plt.figure(figsize=(15,8))
plt.plot(range(1,20), test_error_rates, color='red', marker='o',
         markerfacecolor='yellow', markersize=10)

plt.plot(range(1,20), train_error_rates, color='red', marker='o',
         markerfacecolor='green', markersize=10)

plt.title('Error Rate vs. K Value')
plt.xlabel('K_values')
plt.ylabel('Error Rate')

plt.legend(["Test Error Rate", "Train Error Rate"])
plt.show()
```

```python
# K değerlerini belirliyoruz
K = range(1, 20)

# Eğitim ve doğrulama hatalarını depolayacak listeler
train_error_rates = []
test_error_rates = []

# KNN modelini farklı K değerleri için çalıştırıyoruz
for k in K:
    operations = [("scaler", StandardScaler()), ("knn", KNeighborsClassifier(n_neighbors=k))]
    knn_pipe_model = Pipeline(steps=operations)
    
    knn_pipe_model.fit(X_train, y_train)
    
    train_pred = knn_pipe_model.predict(X_train)
    test_pred = knn_pipe_model.predict(X_test)
    
    train_error = 1 - accuracy_score(y_train, train_pred)
    test_error = 1 - accuracy_score(y_test, test_pred)
    
    train_error_rates.append(train_error)
    test_error_rates.append(test_error)

# Eğitim ve doğrulama hatalarını hesaplıyoruz
train_err = train_error_rates
val_err = test_error_rates

# Grafik oluşturma
fig, ax = plt.subplots(1, 1, figsize=(10, 8))

# Eğitim ve doğrulama hatalarını çiziyoruz
ax.semilogx(K, train_err, 'r-', lw=3, label='Training')
ax.semilogx(K, val_err, 'g-', lw=3, label='Validation')

# Efsane ekliyoruz
ax.legend()

# Eksenleri kontrol ediyoruz
ax.set_xlim(1, 30)
ax.set_ylim(0, 0.3)  
plt.show()
```
